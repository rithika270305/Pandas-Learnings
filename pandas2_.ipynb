```python
students=pd.read_csv("students.csv")
```

```python
students.head()
```

```python
students.merge(nov) # by default:- inner join between tables on common column
```

```python
courses=pd.read_csv("courses.csv")
courses.head()
```

```python
courses.merge(nov,how='left',on='course_id')  #all the rows from left table are present irrespective of match
```

```python
courses.merge(nov,how='right',on='course_id') # opposite to left join can be achieved by reversing the tables
```

```python
students.merge(nov,how='outer',on='student_id') # entire data plus common rows
```

```python
courses.merge(nov,how='inner',on='course_id')['price'].sum() # total revenue in month of november
```

```python
xx=nov.merge(courses,how='inner',on='course_id') # create custom registration_table which contains std_id,name, price,
```

```python
xx
```

```python
registrations_table=xx.merge(students,how='inner',on='student_id')[['student_id','name','price']]
registrations_table.head()
```

```python
result=nov.merge(courses,how='inner',on='course_id').groupby(['course_name'])['price'].sum().sort_values(ascending=False)
result  # total revenue by each subject
```

```python
# reset_index() on series turns it into dataframe where old index becomes new column and new index is added
result=result.iloc[0:3,1].reset_index()
result
```

```python
df = pd.DataFrame({               # the table is in wide format
    'Student': ['A', 'B', 'C'],
    'Math': [90, 80, 70],
    'Science': [85, 75, 65]
})
df
```

```python
df_melt=pd.melt(df,id_vars=['Student'],value_vars=['Science','Math'],value_name='Marks',var_name='Subject')
df_melt
```

```python
df_melt.pivot(index='Student',columns='Subject',values='Marks') # converts long to wide format, only works when there are no duplicate values
```

```python
df_melt.pivot_table(index='Student',columns='Subject',values='Marks',aggfunc='mean') #pivot_table is more powerful than pivot works with nan values and we can also apply agg functions
```

```python
data = {
    "Employee": ["Ravi", "Anu", "Kiran", "Meena", "Arjun"],
    "Join_Date": [
        "2023-05-15",
        "2024-01-20",
        "2022-11-05",
        "2025-03-10",
        "2024-07-25"
    ],
    "Last_Worked": [
        "2024-12-31",
        "2025-06-15",
        "2023-12-01",
        "2025-09-01",
        "2025-01-10"
    ],
    "Salary": [50000, 60000, 55000, 70000, 65000]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
df['Join_Date']=pd.to_datetime(df['Join_Date'])
pd.to_datetime(df['Last_Worked'])
```

```python
df['Year']=df['Join_Date'].dt.year
df['Month']=df['Join_Date'].dt.month
df['day']=df['Join_Date'].dt.day
df['day_name']=df['Join_Date'].dt.day_name()
df['month_name']=df['Join_Date'].dt.month_name()
df['quarter']=df['Join_Date'].dt.quarter
```
