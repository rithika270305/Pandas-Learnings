**4.Pandas DataFrames**

```python
student=[
    [100,80,50],
    [56,78,65],
    [55,23,99],
    [89,92,94]

]
pd.DataFrame(student,columns=['English','Math','Social'],index=['A','B','C','D'])
```

```python
student2={
    'iq':[56,78,65,89],
    'marks':[55,23,99,92]
}
student2=pd.DataFrame(student2,index=['A','B','C','D'])
student2
```

```python
print(student2.loc["B":, :]) #indexing in pandas does not use direct slicing with labels instead .loc(index based labels), .iloc(integer based)
```
